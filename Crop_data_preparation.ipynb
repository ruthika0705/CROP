```python
import pandas as pd
import numpy as np
```

```python
def final(min1,max1,min2,max2,min3,max3,label):
    temperature = np.random.uniform(min1, max1,100)
    humidity = np.random.uniform(min2, max2,100)
    rainfall = np.random.uniform(min3, max3,100)
    #labels ={label*100}
    series1 =pd.Series(temperature, name = "temperature")
    series2 =pd.Series(humidity, name="humidity")
    series3 =pd.Series(rainfall, name="rainfall")
    #df = pd.DataFrame(series1,series2,series3)
    series1 = pd.DataFrame(series1)
    df = series1.join(series2)
    df1 = df.join(series3)
    df1["label"] = label
    #print(df1.head())
    df1.to_csv("data_22-03-2021.csv", mode='a', header=False)
    #print(df1)
    return df1
    
```
